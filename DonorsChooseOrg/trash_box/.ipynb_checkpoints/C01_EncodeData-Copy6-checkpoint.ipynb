```python
project_submitted_datetime_year = pd.to_datetime(train_df["project_submitted_datetime"]).dt.year
project_submitted_datetime_year = project_submitted_datetime_year.rename('project_submitted_datetime_year', axis='columns')

project_submitted_datetime_month = pd.to_datetime(train_df["project_submitted_datetime"]).dt.month
project_submitted_datetime_month = project_submitted_datetime_month.rename('project_submitted_datetime_month', axis='columns')

project_submitted_datetime_month = pd.to_datetime(train_df["project_submitted_datetime"]).dt.month
project_submitted_datetime_month = project_submitted_datetime_month.rename('project_submitted_datetime_month', axis='columns')

project_submitted_datetime_day = pd.to_datetime(train_df["project_submitted_datetime"]).dt.day
project_submitted_datetime_day = project_submitted_datetime_day.rename('project_submitted_datetime_day', axis='columns')

project_submitted_datetime_dayofweek = pd.to_datetime(train_df["project_submitted_datetime"]).dt.dayofweek
project_submitted_datetime_dayofweek = project_submitted_datetime_dayofweek.rename('project_submitted_datetime_dayofweek', axis='columns')

train_df = pd.concat([train_df, project_submitted_datetime_year, project_submitted_datetime_month, project_submitted_datetime_day, project_submitted_datetime_dayofweek], axis=1) 
train_df
```
