```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.model_selection import cross_validate
import tensorflow as tf
```

```python
train_df = pd.read_csv("./C01_train_encoded.csv")
#train_df.info()
```

```python
train_df.head(10)
```

```python
train_data = train_df.values
xs = train_data[:, 2:]  # id,正解データ列以外を抽出
y = train_data[:, 1]  # 正解データ列のみ抽出
```

```python
xs.shape
```

```python
y.shape
```

```python
y[y[0,] == 1].sum() / y.size
```

```python
train_X = xs[:150000, :]
train_Y = y[:150000]

test_X = xs[150000:, :]
test_Y = y[150000:]
```

```python
train_X.shape
```

```python
test_X.shape
```

```python

#model = tf.keras.models.Sequential([
#  tf.keras.layers.Flatten(),
#  tf.keras.layers.Dense(32, activation=tf.nn.relu, input_shape=82),
#  tf.keras.layers.Dropout(0.2),

#  tf.keras.layers.Dense(10, activation=tf.nn.relu), # あってもなくても同じ(2秒遅くなる)
#  tf.keras.layers.Dropout(0.2),

#  tf.keras.layers.Dense(2, activation=tf.nn.softmax)
#  tf.keras.layers.Dense(1, activation=tf.nn.softmax)
#])

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(1024, input_dim=xs.shape[1], activation='sigmoid'))
model.add(Dropout(0.5))
model.add(Dense(512, activation='sigmoid'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam',
#              loss='sparse_categorical_crossentropy',
#              loss='categorical_crossentropy',
              loss='binary_crossentropy',
              metrics=['accuracy'])

#model.fit(xs, y, epochs=5)
#loss, accuracy = model.evaluate(xs, y)

model.fit(train_X, train_Y, epochs=15)
loss, accuracy = model.evaluate(test_X, test_Y)

print("loss = {:.2f}".format(loss))
print("acc = {:.2f}".format(accuracy))
```
